## Finally displaying the data of the companies

```python
import os
import pandas as pd
datasets=os.listdir("C:\\Users\\kavin\\OneDrive\\Desktop\\Data ritz ML work\\Myproject\\Data-ritz-project\\Datasets\\HISTORICAL_DATA\\HISTORICAL_DATA1\\")
c=input("Enter the Company name :")
c=c.upper()
c=list(c)
for i in range(len(c)):
    if c[i]==' ':
        c[i]='_'
c=''.join(c)
c=c+"_data.csv"
for i in datasets:
    if i==c:
        path="C:\\Users\\kavin\\OneDrive\\Desktop\\Data ritz ML work\\Myproject\\Data-ritz-project\\Datasets\\HISTORICAL_DATA\\HISTORICAL_DATA1\\"+i
        dataset=pd.read_csv(path)
        if len(dataset)<200:
            print("Insufficient Data")
            break
        else:
            dataset=dataset.dropna(how="any",subset=['open','high','volume','close'])
            dataset=dataset.drop_duplicates(keep='last')
print(dataset)
```
