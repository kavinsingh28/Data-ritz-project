## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
A=np.linspace(0,100,1000)
%matplotlib inline
x=A
y=np.sin(A)
```

```python
# k is the color code for black color
plt.plot(x,y,color='k',linestyle="dashdot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sinusoidal graph")
plt.show()
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
X=["Delhi","Mumbai","Chennai","Kolkata","Hyderabad"]
Clothing=np.array([82,78,48,46,39])
Equipment=np.array([48,31,51,29,18])
Accessories=np.array([15,20,14,13,9])
A=[i for i in range(1,len(X)+1)]
plt.bar(A,Clothing,width=0.8,color='red',label='Clothing',bottom=Equipment+Accessories)
plt.bar(A,Equipment,width=0.8,color='silver',label='Equipment',bottom=Accessories)
plt.bar(A,Accessories,width=0.8,color='green',label='Accessories')

plt.xlabel("Cities",size=14)
plt.ylabel("Total Stores",size=14)
plt.xticks(A,X,rotation=90)
plt.title("City wise data of shopping stores",size=20)
plt.legend(loc=0)
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
A=sb.load_dataset('titanic')
A.head()
```

```python
X=A['pclass']
Y=A['survived']
color=['Blue','Orange','Green']
sb.barplot(X,Y,hue='who',data=A,palette=color)
plt.xlabel("Passenger Class",size=14)
plt.ylabel("Survival Rate",size=14)
plt.title("Survival data according to different Age",size=20)
plt.legend(loc=0)
plt.show()
```
